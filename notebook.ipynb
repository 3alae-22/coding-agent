```python
from dotenv import load_dotenv
load_dotenv()

import anthropic 
anthropic_client=anthropic.Anthropic()
```

```python
system_prompt = "You are a very professional old player of soccer and a very smart coach."
user_prompt = "Tell me who will win the word cup and the first 4 teams."

chat_messages = [
    {"role": "user", "content": user_prompt}
]

response = anthropic_client.messages.create(
    model='claude-haiku-4-5-20251001',
    system=system_prompt,
    messages=chat_messages,
    max_tokens=1024
)

print(response.content[0].text)
```

```python
import random
def make_joke(name):
    jokes = [
        f"Why did {name} bring a pencil to the party? Because he wanted to draw some attention!",
        f"Did you hear about {name}'s bakery? Business is on a roll!",
        f"{name} walked into a library and asked for a burger. The librarian said, 'This is a library.' So {name} whispered, 'Can I get a burger?'",
        f"When {name} does push-ups, the Earth moves down.",
        f"{name} told a chemistry joke... but there was no reaction.",
    ]
    return random.choice(jokes)
```

```python
make_joke_description = {
    "name": "make_joke",
    "description": "Generates a random personalized joke using the provided name. Use when the user asks for a joke or something funny.",
    "input_schema": {
        "type": "object",
        "properties": {
            "name": {
                "type": "string",
                "description": "The name to insert into the joke, personalizing the output.",
            }
        },
        "required": ["name"],
    },
}
```

```python
system_prompt = "You can make funny and original jokes. Find out the user's name to make the joke personalized."
user_prompt = "Tell me a joke about Alexey."

chat_messages = [
    {"role": "user", "content": user_prompt}
]

response = anthropic_client.messages.create(
    model='claude-haiku-4-5-20251001',
    system=system_prompt,
    messages=chat_messages,
    tools=[make_joke_description],
    max_tokens=1024
)

print(response)
```

```python

from anthropic import Anthropic

from toyaikit.tools import Tools
from toyaikit.chat import IPythonChatInterface
from toyaikit.llm import AnthropicClient
from toyaikit.chat.runners import AnthropicMessagesRunner
```

```python
tools_obj = Tools()
tools_obj.add_tools(make_joke_description)
```

```python
chat_interface = IPythonChatInterface()
llm_client = AnthropicClient(model="claude-sonnet-4-6")

runner = AnthropicMessagesRunner(
    tools=tools_obj,
    developer_prompt=system_prompt,
    chat_interface=chat_interface,
    llm_client=llm_client
)

runner.run()
```
